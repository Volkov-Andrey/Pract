```python
from random import randint
letters, cells = 'ABC', [0]*9
for i in range(9):
    cells[i] = randint(1,9)
def pg():
    print(' ','---'*4)
    for i in range (0,9,3):
        print (str(int(i/3+1)),'|', cells[i], '|', cells[i+1], '|', cells[i+2], '|')
        print(' ','---'*4)
    print ('   ', 'A', ' ', 'B', ' ','C')
    return ('')
stroke, column = {}, {}
for i in range(3):
    for j in range(3):
        if str(i+1) not in stroke:
            stroke[str(i+1)] = [i*3+j]
        else:
            stroke[str(i+1)].append(i*3+j)
for i in range(len(letters)):
    for j in range(3):
        if letters[i] not in column:
            column[letters[i]] = [i+j*3]
        else:
            column[letters[i]].append(i+j*3)
def turn(act):
    crd = 0
    for i in stroke[act[1]]:
        for j in column[act[0]]:
            if i == j:
                crd = i
                return (crd)
            else:
                pass
def check_correct(act,last_act,prt):
    flag = False
    if prt == 0 or last_act == '':
        return True
    if prt % 2 == 1:
        for i in column:
            cur_column = column[i]
            if turn(act) in cur_column and turn(last_act) in cur_column:
                flag = True
    if prt % 2 == 0:
        for i in stroke:
            cur_stroke = stroke[i]
            if turn(act) in cur_stroke and turn(last_act) in cur_stroke:
                flag = True
    return flag
def attempts(act,prt):
    check = ''
    if prt == 0:
        return True
    if prt % 2 == 1:
        for i in stroke[act[1]]:
            check += str(cells[i])
        if check == '   ':
            return False
        else:
            return True
    if prt % 2 == 0:
        for i in column[act[0]]:
            check += str(cells[i])
        if check == '   ':
            return False
        else:
            return True
def possible_to_play_check():
    check_stroke = ''
    for i in cells:
        check_stroke += str(i)
    if check_stroke == '         ':
        return True
    else:
        return False
def player_act():
    p1, p2, prt, last_act = 0, 0, 0, ''
    print(pg())
    while True:
        if possible_to_play_check():
            break
        if prt % 2 == 0:
            print('Ход первого игрока:')
            act = input('Введите комбинацию A-C + 1-3, например: B2 ')
        if prt % 2 == 1:
            print('Ход второго игрока:')
            act = input('Введите комбинацию A-C + 1-3, например: B2 ')
        if act[0] not in column or act[1] not in stroke or len(act) != 2:
            print('Неверный формат ввода!')
            continue
        if cells[turn(act)] == ' ':
            print('Выберите то поле, в котором есть число!')
            continue
        if not check_correct(act, last_act, prt):
            print('Такой ход невозможен!')
            continue
        if prt % 2 == 0:
            p1 += cells[turn(act)]
            cells[turn(act)] = ' '
            print(pg())
            last_act = act
            if attempts(act, prt):
                prt += 1
            print(f'У первого игрока:{p1}\nУ второго игрока:{p2}')
            continue
        if prt % 2 == 1:
            p2 += cells[turn(act)]
            cells[turn(act)] = ' '
            print(pg())
            last_act = act
            if attempts(act, prt):
                prt += 1
            print(f'У первого игрока:{p1}\nУ второго игрока:{p2}')
            continue
    if p1 > p2:
        return (f'Победил первый игрок!')
    if p1 < p2:
        return (f'Победил второй игрок!')
    if p1 == p2:
        return (f'Ничья!')
def game():
    print ('Нажмите Enter чтобы начать!')
    enter=input()
    print(player_act())
print(game())

```
