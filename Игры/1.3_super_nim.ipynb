```python
import random
digits = [' ']*64
bukv = 'ABCDEFGH'
stroke, row = {}, {}
for i in range(random.randint(16,30)):
    digits[random.randint(0,63)] = '*'
def pg():
    print(' ','--'*17)
    for i in range(0,64,8):
        print(str(int((64-i)/8)),'|',digits[i],'|',digits[i+1],'|',digits[i+2],'|',digits[i+3],'|',digits[i+4],'|',digits[i+5],'|',digits[i+6],'|',digits[i+7],'|')
        print(' ','--'*17)
    print('   ','A',' ','B',' ','C',' ','D',' ','E',' ','F',' ','G',' ','H')
    return ''
for i in range(1,9):
    for j in range(8):
        if str(9-i) not in stroke:
            stroke[str(9-i)] = [8*(i-1)+j]
            continue
        if str(9-i) in stroke:
            stroke[str(9-i)].append(8*(i-1)+j)
for i in range(len(bukv)):
    for j in range(8):
        if bukv[i] not in row:
            row[bukv[i]] = [i+j*8]
        else:
            row[bukv[i]].append(i+j*8)
def pa():
    player = 0
    while True:
        if '*' not in digits:
            break
        print('Введите A-H или 1-8:')
        move = str(input())
        counter = 0
        if move =='1' or move =='2' or move =='3' or move =='4' or move =='5' or move =='6' or move =='7' or move =='8':
            for i in stroke[move]:
                if digits[i] == '*':
                    counter +=1
                    digits[i] = ' '
            if counter == 0:
                print('В этой строке не осталось точек, ведите другое значение')
                continue
            else:
                print(pg())
        elif move =='A' or move =='B' or move =='C' or move =='D' or move =='E' or move =='F' or move =='G' or move =='H':
            for i in row[move]:
                    if digits[i] == '*':
                        counter +=1
                        digits[i] = ' '
            if counter == 0:
                print('В этом столбце не осталось точек, введите другое значение')
                continue
            else:
                print(pg())
        elif move !='1' or move !='2' or move !='3' or move !='4' or move !='5' or move !='6' or move !='7' or move !='8' or\
            move !='A' or move !='B' or move !='C' or move !='D' or move !='E' or move !='F' or move !='G' or move !='H':
            print('Некорректен формат ввода! Введите либо 1-8 либо A-H')
            continue
        player += 1
    if player % 2 == 0:
        return ('Поздравляем! Выиграл игрок "2"!')
    else:
        return ('Поздравляем! Выиграл игрок "1"!')
def game():
    print('Нажмите Enter')
    enter=input()
    print(pg())
    print(pa())
    return ('')
print(game())
```
